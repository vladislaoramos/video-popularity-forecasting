```python
ATTENTION_LINES = [
    '!', '(', ')', '[', ']', '|', '$', '#', '&', 'official', 'Official', '–ü–†–ï–ú–¨–ï–†–ê', '–ø—Ä–µ–º—å–µ—Ä–∞', '–ü—Ä–µ–º—å–µ—Ä–∞', '2021', '#',
    'how to', 'movie', 'movies', 'Official Music Video', 'Official Video'
]
```

```python
def checker(title):
    for item in ATTENTION_LINES:
        if item in title:
            return True
```

```python
data["attention_flag"] = data.title.map(lambda x: 1 if checker(x) else 0)
```

```python
import datetime


def how_many_days(video_date, channel_date):
    # today = datetime.datetime.today()
    vid_date = datetime.datetime.fromisoformat(video_date[:10])
    ch_date = datetime.datetime.fromisoformat(channel_date[:10])
    return (vid_date - ch_date).days
```

```python
data["ch_age_before_vid"] = data.apply(lambda x: how_many_days(x["video_published_at"], x["channel_published_at"]), axis=1)
```
